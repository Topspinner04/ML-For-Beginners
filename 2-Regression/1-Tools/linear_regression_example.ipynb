```python
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model, model_selection
```

Imports relevant modules of package scikit-learn. To visualize I use matplotlib.

```python
housing = datasets.fetch_california_housing()

print(housing.data.shape)
print(housing.target.shape)
print(housing.feature_names)
print(housing.target_names)
```

We can see that the housing data contains 20640 rows and eight columns, being the medium income, median house age, average number of rooms, average number of bedrooms, population of the block, average number of household members, the latitude and the longitude. The target variable is the median house value, expressed in hundreds of thousands of dollars.

```python
X, y = datasets.fetch_california_housing(return_X_y=True)
X = X[:,0]
X = X.reshape((-1, 1))
```

We will load the data into X and the target into y. We will only perform the linear regression on the first column of the data, being the median income.

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

We will now split the data into test data and training data. We use the linear regression model from scikit-learn, and train the model using the training data.

```python
y_pred = model.predict(X_test)
```

Now we will predict the target variable using the test data. The model will assign a specific target (median house value) to every data point in our testing data (median income).

```python
plt.scatter(X_test, y_test, color="black")
plt.plot(X_test, y_pred, color="Blue", linewidth=3)
plt.xlabel("Median Income")
plt.ylabel("Median House Value")
plt.title("A Graph Plot Showing Median House Value Against Median Income")
plt.show()
```

We can see that the higher the median income, the more valuable is the median house value. Note: I do not know why the median house value caps at 5 (or 500.000$).
